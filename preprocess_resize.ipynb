```python
import pandas as pd
import numpy as np
from PIL import ImageFile, Image
import os

ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
df_data = pd.read_csv(f"{path}/2012_2019_PlatteRiverWeir_features_merged_all.csv")
df_data.head()
```

```python
def resize_image(path, file, new_path):
  image = Image.open(f"{path}/{file}")

  #print(f"Original size : {image.size}")
  image_resized = image.resize((1024, 1024))
  image_resized.save(f"{new_path}/{file}")
```

```python
filenames = df_data.Filename
filenames = [file.lower() for file in filenames]
filenames = set(filenames)
```

```python
new_path = 'drive/MyDrive/project_hydrology_images/preprocess_image_resize_bigger'
```


Check files that already exist in new path

```python
existing_files = set()

for root, dirs, files in os.walk(new_path):
    path_images = root.split(os.sep)
    print((len(path_images) - 1) * '---', os.path.basename(root))
    for file in files:
        #print(file)
        existing_files.add(file.lower())
```

```python
print(existing_files)
```

```python
for root, dirs, files in os.walk(path_images):
    path_images = root.split(os.sep)
    print((len(path_images) - 1) * '---', os.path.basename(root))

    for file in files:
        if file.lower() in filenames and file.lower() not in existing_files:
          resize_image('/'.join(path_images), file, new_path)
```

```python
for root, dirs, files in os.walk(path_images):
    path_images = root.split(os.sep)
    print((len(path_images) - 1) * '---', os.path.basename(root))
    for file in files:
        file_size = os.path.getsize(f"{'/'.join(path_images)}/{file}")
        if file_size == 0:
          print(file)
```
